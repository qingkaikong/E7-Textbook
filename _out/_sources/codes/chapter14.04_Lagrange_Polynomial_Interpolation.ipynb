```python
import numpy as np
import numpy.polynomial.polynomial as poly
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# import 
X = np.array([0,1,2], dtype=float)
Y = np.array([1,3,2], dtype=float)

P1 = np.poly1d([.5,-1.5,1])
P2 = np.poly1d([-1,2,0])
P3 = np.poly1d([.5,-.5,0])

x = np.arange(-1.0, 3.0, 0.1)

fig = plt.figure()
plt.plot(x, np.polyval(P1,x), 'b',
            x, np.polyval(P2,x), 'r',
            x, np.polyval(P3,x), 'g');
plt.plot(X, np.ones(X.shape), 'ko',
            X, np.zeros(X.shape), 'ko');
plt.grid()
plt.title('Lagrange Basis Polynomials');
plt.xlabel('x');
plt.ylabel('y');
```

```python
L = P1 + 3*P2 + 2*P3

fig = plt.figure()
plt.plot(X, Y, 'ro', x, L(x), 'b')
plt.title('Lagrange Polynomial');
plt.grid()
plt.xlabel('x');
plt.ylabel('y');
```
