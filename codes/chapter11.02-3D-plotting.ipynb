## __11.2__ 3D Plotting

In order to plot on 3-dimensional axes, we import the `mplot3d` package:

```python
from mpl_toolkits import mplot3d
```

Once we have imported this submodule, we can create axes with the following command:

```python
ax = plt.axes(projection='3d')
```

The most basic 3D plot is from sets of $(x, y, z)$ triples. We use the `ax.plot3D` function to plot these sets of triples.

__EXAMPLE 11.9: Basic 3-Dimensional Plot__

Consider the parameterized data set from $t = 0$ to $t = 10\pi$, $x = sin(t)$ and $y = cos(t)$. We construct a 3-D plot from this data.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
ax.grid()
t = np.linspace(0, 10 * pi, 500)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t)
plt.title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('t')
```

Many times we would like a surface plot rather than a line plot when plotting in three dimensions. In three-dimensional surface plotting, we wish to make a graph of some relationship $f(x, y)$. In surface plotting all $(x,y)$ pairs must be given. This is not straightforward to do using vectors. Therefore, in
surface plotting, the first data structure you must create is called a __mesh__. Given vectors of $x$ and $y$ values, a mesh is a listing of all the possible combinations of x and y. In Python, the mesh is given as two
matrices `X` and `Y` where `X[i, j]` and `Y[i, j]` define possible $(x,y)$ pairs. A third matrix, `Z`, can then be created such that `Z[i, j] = f(X [i,j], Y[i,j])`. A mesh can be created using the `meshgrid` function in `numpy`. 

__EXAMPLE 11.10: Meshgrid Creation__

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d

xdata = np.array([1, 2, 3, 4])
ydata = np.array([3, 4, 5])

[X, Y] = np.meshgrid(xdata, ydata)
```

To plot a surface, we can directly plot a `meshgrid` using several different functions. 

__EXAMPLE 11.11: Surface Plotting Using `plot_surface`__

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d

xdata = np.arange(-5, 5.2, 0.2)
ydata = np.arange(-5, 5.2, 0.2)
[X, Y] = np.meshgrid(xdata, ydata)

Z = np.sin(X) * np.cos(Y)

ax = plt.axes(projection='3d')
ax.grid()
ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

__EXAMPLE 11.12: Contour Plotting__

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d

xdata = np.arange(-5, 5.2, 0.2)
ydata = np.arange(-5, 5.2, 0.2)
[X, Y] = np.meshgrid(xdata, ydata)
Z = np.sin(X) * np.cos(Y)

plt.contour(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
```

A final example that can be explored is the `quiver` function to plot a 3D field of arrows. This function takes in the usual matrices `X`, `Y`, and `Z`, and also three additional arguments `U`, `V`, and `W` that specify the length of the arrows in each Cartesian direction.

__EXAMPLE 11.13: Quiver Plotting__

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d

xdata = np.arange(-5, 5.2, 1)
ydata = np.arange(-5, 5.2, 1)
[X, Y] = np.meshgrid(xdata, ydata)

Z = np.sin(X) * np.cos(Y)

ax = plt.axes(projection='3d')
ax.grid()
ax.quiver(X, Y, Z, 0.3, 0.3, 0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
