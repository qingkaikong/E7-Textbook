## __11.2__ 3D Plotting

In order to plot on 3-dimensional axes, we import the `mplot3d` package:

```python
from mpl_toolkits import mplot3d
```

Once we have imported this submodule, we can create axes with the following command:

```python
ax = plt.axes(projection='3d')
```

The most basic 3D plot is from sets of $(x, y, z)$ triples. We use the `ax.plot3D` function to plot these sets of triples.

__EXAMPLE 11.9: Basic 3-Dimensional Plot__

Consider the parameterized data set from $t = 0$ to $t = 10\pi$, $x = sin(t)$ and $y = cos(t)$. We construct a 3-D plot from this data.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
ax.grid()
t = np.linspace(0, 10 * pi, 500)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t)
plt.title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('t')
```
