## __11.1__ 2D Plotting

The basic plotting function in Python comes from the `matplotlib.pyplot` library. We can call a `plot(x, y)` function that can provide a clean visual display of two arrays of data, `x` and `y`.

__EXAMPLE 11.1: Basic Plot__

```python
import numpy as np
import matplotlib.pyplot as plt

# Example plot of x against y
xdata = np.array([0, 1, 2, 3])
ydata = np.power(xdata, 2)
plt.plot(xdata, ydata) # Plots x and y
```

You will notice that this function connects each point with a blue line. To make the function look more "smooth", we should use a finer discretization of points. See the following example for a smoother example of the previous curve:

__EXAMPLE 11.2: Smooth Plot__

```python
import numpy as np
import matplotlib.pyplot as plt

xdata = np.linspace(0, 3, 100) # 100 points instead of 3!
ydata = np.power(xdata, 2)
plt.plot(xdata, ydata) # Smooth!
```

__EXAMPLE 11.3: Plot Customization__

```python
import numpy as np
import matplotlib.pyplot as plt

xdata = np.linspace(-5, 5, 100)
ydata = np.power(xdata, 3) # Plot f(x) = x^3
plt.plot(xdata, ydata, 'g--') # Green dotted line!
```

It is customary in engineering to _always_ give your plot a title and axis labels. The `plt.title` function is used for the plot title, and the `plt.xlabel` and `plt.ylabel` functions are used for axes labels.

__EXAMPLE 11.4: Plot Customization with Titles__

```python
import numpy as np
import matplotlib.pyplot as plt

xdata = np.linspace(-5, 5, 100)
squared = np.power(xdata, 2) # Plot f(x) = x^2
cubed = np.power(xdata, 3) # Plot f(x) = x^3
plt.plot(xdata, squared, 'r+-') # + marks as points, red solid line
plt.plot(xdata, cubed, 'ko-') # circles as points, black solid line
plt.title('Plot of Various Polynomials')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```
